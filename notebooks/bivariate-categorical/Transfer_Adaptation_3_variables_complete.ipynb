```python
%matplotlib inline
import numpy as np
import torch
import matplotlib.pyplot as plt
from tqdm import tnrange, tqdm_notebook

from causal_meta.utils.data_utils import generate_data_categorical_3_variables, generate_data_categorical_3_variables_dependency
from causal_meta.bivariate.categorical1 import StructuralModel
```

```python
N = 10
```

```python
num_episodes = 500
batch_size = 100 # 1
num_test = 10000
num_training = 10 # 100
num_transfers = 10 # 100
```

```python
model = StructuralModel(N) #, dtype=torch.float64)
optimizer = torch.optim.SGD(model.modules_parameters(), lr=1.0)
```

```python
losses1 = np.zeros((2, num_training, num_transfers, num_episodes))
losses2 = np.zeros((2, num_training, num_transfers, num_episodes))
losses3 = np.zeros((2, num_training, num_transfers, num_episodes))
losses4 = np.zeros((2, num_training, num_transfers, num_episodes))
losses5 = np.zeros((2, num_training, num_transfers, num_episodes))
losses6 = np.zeros((2, num_training, num_transfers, num_episodes))

for k in tnrange(num_training):
    
    #Marginal Probability
    pi_A_1 = np.random.dirichlet(np.ones(N))
    
    #Conditional Probability
    
    pi_B_A = np.random.dirichlet(np.ones(N), size=N)
    pi_C_B = np.random.dirichlet(np.ones(N), size=N)

    for j in tnrange(num_transfers, leave=False):
        
        pi_A_2 = np.random.dirichlet(np.ones(N))
                         
        x_val = torch.from_numpy(generate_data_categorical_3_variables(num_test, pi_A_2, pi_B_A,  pi_C_B))
        
        for i in range(num_episodes):
            
            x_transfer = torch.from_numpy(generate_data_categorical_3_variables(batch_size, pi_A_2, pi_B_A, pi_C_B))
          
            model.zero_grad()
            
            
            #Model A-->B-->C
            #Model A<--B<--C
            
            loss_A_B_C = -torch.mean(model.model_A_B_C(x_transfer))
            loss_C_B_A = -torch.mean(model.model_C_B_A(x_transfer))
            
            #Model B-->A-->C
            #Model B<--A<--C
                        
            loss_B_A_C = -torch.mean(model.model_B_A_C(x_transfer))
            loss_C_A_B = -torch.mean(model.model_C_A_B(x_transfer))
            
            #Model A-->C-->B
            #Model A<--C<--B
            
            loss_A_C_B = -torch.mean(model.model_A_C_B(x_transfer))
            loss_B_C_A = -torch.mean(model.model_B_C_A(x_transfer))
            
            #Model B<--A-->C
            #Model B-->A<--C
            
            loss_A_B_C_1 = -torch.mean(model.model_A_B_C_1(x_transfer))
            loss_A_C_B_1 = -torch.mean(model.model_C_B_A_1(x_transfer))
            
            #Model A<--C-->B
            #Model A-->C<--B
                        
            loss_C_A_B_1 = -torch.mean(model.model_B_A_C_1(x_transfer))
            loss_C_B_A_1 = -torch.mean(model.model_C_A_B_1(x_transfer))
            
            #Model C<--B-->A
            #Model C-->B<--A
            
            loss_B_A_C_1 = -torch.mean(model.model_A_C_B_1(x_transfer))
            loss_B_C_A_1 = -torch.mean(model.model_B_C_A_1(x_transfer))
            
            
            loss = loss_A_B_C + loss_A_C_B + loss_C_A_B + loss_C_B_A + loss_B_A_C + loss_B_C_A + loss_A_B_C_1 + loss_C_B_A_1 + loss_B_A_C_1 + loss_C_A_B_1 + loss_A_C_B_1 + loss_B_C_A_1
            
            with torch.no_grad():
                
                val_loss_A_B_C = -torch.mean(model.model_A_B_C(x_val))
                val_loss_C_B_A = -torch.mean(model.model_C_B_A(x_val))
                
                val_loss_B_A_C = -torch.mean(model.model_B_A_C(x_val))
                val_loss_C_A_B = -torch.mean(model.model_C_A_B(x_val))
                
                val_loss_A_C_B = -torch.mean(model.model_A_C_B(x_val))
                val_loss_B_C_A = -torch.mean(model.model_B_C_A(x_val))
                
                val_loss_A_B_C_1 = -torch.mean(model.model_A_B_C_1(x_val))
                val_loss_C_B_A_1 = -torch.mean(model.model_C_B_A_1(x_val))
                
                val_loss_B_A_C_1 = -torch.mean(model.model_B_A_C_1(x_val))
                val_loss_C_A_B_1 = -torch.mean(model.model_C_A_B_1(x_val))
                
                val_loss_A_C_B_1 = -torch.mean(model.model_A_C_B_1(x_val))
                val_loss_B_C_A_1 = -torch.mean(model.model_B_C_A_1(x_val))

            losses1[:, k, j, i] = [val_loss_A_B_C.item(), val_loss_C_B_A.item()]
            losses2[:, k, j, i] = [val_loss_B_A_C.item(), val_loss_C_A_B.item()]
            losses3[:, k, j, i] = [val_loss_A_C_B.item(), val_loss_B_C_A.item()]
            
            losses4[:, k, j, i] = [val_loss_A_B_C_1.item(), val_loss_C_B_A_1.item()]
            losses5[:, k, j, i] = [val_loss_B_A_C_1.item(), val_loss_C_A_B_1.item()]
            losses6[:, k, j, i] = [val_loss_A_C_B_1.item(), val_loss_B_C_A_1.item()]

            loss.backward()
            optimizer.step()
            
            
```

```python
flat_losses1 = -losses1.reshape((2, -1, num_episodes))
losses1_25, losses1_50, losses1_75 = np.percentile(flat_losses1, (25, 50, 75), axis=1)

flat_losses2 = -losses2.reshape((2, -1, num_episodes))
losses2_25, losses2_50, losses2_75 = np.percentile(flat_losses2, (25, 50, 75), axis=1)

flat_losses3 = -losses3.reshape((2, -1, num_episodes))
losses3_25, losses3_50, losses3_75 = np.percentile(flat_losses3, (25, 50, 75), axis=1)

flat_losses4 = -losses4.reshape((2, -1, num_episodes))
losses4_25, losses4_50, losses4_75 = np.percentile(flat_losses4, (25, 50, 75), axis=1)

flat_losses5 = -losses5.reshape((2, -1, num_episodes))
losses5_25, losses5_50, losses5_75 = np.percentile(flat_losses5, (25, 50, 75), axis=1)

flat_losses6 = -losses6.reshape((2, -1, num_episodes))
losses6_25, losses6_50, losses6_75 = np.percentile(flat_losses6, (25, 50, 75), axis=1)

plt.figure(figsize=(14, 9))
ax = plt.subplot(1, 1, 1)

ax.plot(losses1_50[0], color='C0', label=r'$A \rightarrow B \rightarrow C$', lw=2)
ax.fill_between(np.arange(num_episodes), losses1_25[0], losses1_75[0], color='C0', alpha=0.2)

ax.plot(losses1_50[1], color='C3', label=r'$A \leftarrow B \leftarrow C$', lw=2)
ax.fill_between(np.arange(num_episodes), losses1_25[1], losses1_75[1], color='C3', alpha=0.2)

ax.plot(losses2_50[0], color='C5', label=r'$A \rightarrow C \rightarrow B$', lw=2)
ax.fill_between(np.arange(num_episodes), losses2_25[0], losses2_75[0], color='C5', alpha=0.2)

ax.plot(losses2_50[1], color='C7', label=r'$A \leftarrow C \leftarrow B$', lw=2)
ax.fill_between(np.arange(num_episodes), losses2_25[1], losses2_75[1], color='C7', alpha=0.2)

ax.plot(losses3_50[0], color='C9', label=r'$C \rightarrow B \rightarrow A$', lw=2)
ax.fill_between(np.arange(num_episodes), losses3_25[0], losses3_75[0], color='C9', alpha=0.2)

ax.plot(losses3_50[1], color='C11', label=r'$C \leftarrow B \leftarrow A$', lw=2)
ax.fill_between(np.arange(num_episodes), losses3_25[1], losses3_75[1], color='C11', alpha=0.2)

ax.plot(losses4_50[0], color='C1', label=r'$A \leftarrow B \rightarrow C$', lw=2)
ax.fill_between(np.arange(num_episodes), losses4_25[0], losses4_75[0], color='C1', alpha=0.2)

ax.plot(losses4_50[1], color='C2', label=r'$C \rightarrow B \leftarrow A$', lw=2)
ax.fill_between(np.arange(num_episodes), losses4_25[1], losses4_75[1], color='C2', alpha=0.2)

ax.plot(losses5_50[0], color='C4', label=r'$B \leftarrow A \rightarrow C$', lw=2)
ax.fill_between(np.arange(num_episodes), losses5_25[0], losses5_75[0], color='C4', alpha=0.2)

ax.plot(losses5_50[1], color='C6', label=r'$C \rightarrow A \leftarrow B$', lw=2)
ax.fill_between(np.arange(num_episodes), losses5_25[1], losses5_75[1], color='C6', alpha=0.2)

ax.plot(losses6_50[0], color='C8', label=r'$A \leftarrow C \rightarrow B$', lw=2)
ax.fill_between(np.arange(num_episodes), losses6_25[0], losses6_75[0], color='C8', alpha=0.2)

ax.plot(losses6_50[1], color='C10', label=r'$B \rightarrow C \leftarrow A$', lw=2)
ax.fill_between(np.arange(num_episodes), losses6_25[1], losses6_75[1], color='C10', alpha=0.2)

ax.set_xlim([0, flat_losses1.shape[1] - 1])
ax.tick_params(axis='both', which='major', labelsize=13)
ax.legend(loc=4, prop={'size': 13})
ax.set_xlabel('Number of examples', fontsize=14)
ax.set_ylabel(r'$\log P(D\mid \cdot \rightarrow \cdot)$', fontsize=14)

plt.show()
```
