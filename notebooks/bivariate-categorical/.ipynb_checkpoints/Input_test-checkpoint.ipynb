```python
import numpy as np

s = np.random.dirichlet((10, 5, 3,2), 20).transpose()
```

```python
import matplotlib.pyplot as plt
plt.barh(range(20), s[0])
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import splev, splrep

x = np.linspace(0, 10, 10)
y = np.sin(x)
spl = splrep(x, y)
x2 = np.linspace(0, 10, 200)
y2 = splev(x2, spl)

plt.plot(x, y, 'o')
plt.plot(x2+1, y2)
plt.show()
```
